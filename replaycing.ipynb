```python
cars_df['Luggage.room']=cars_df['Luggage.room'].fillna(0)
cars_df['Luggage.room'].isnull().sum()
```

                    Replyacing with Mean

```python
cars_df['Price']=cars_df['Price'].fillna(cars_df['Price'].mean())
cars_df['Price']
```

                  Replyacing with Mode

```python
cars_df['Luggage.room']=cars_df['Luggage.room'].fillna(cars_df['Luggage.room'].mode()[0])
cars_df['Luggage.room']
```

                  Replyacing with Median

```python
cars_df['Max.Price']=cars_df['Max.Price'].fillna(cars_df['Max.Price'].median())
cars_df['Max.Price']
```
