```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('D:\\exam1\\mtcars 1.csv')
df
```

```python
range=(10,40)
bins=10
plt.hist(df['mpg'],bins,range,color="red",rwidth=0.8)
plt.xlabel('mpg')
plt.ylabel('no of frequency')
plt.title('my histogram')
plt.show()
```

```python
plt.scatter(df['wt'],df['mpg'],color='black',marker="*")
plt.xlabel('wt')
plt.ylabel('mpg')
plt.title('my scatter plot')
plt.show()
```

```python
temp=df['am'],value_count()
print(temp)
temp.plt.bar
plt.xlabel('am')
plt.ylabel('no of frequency transmission')
plt.title('my bar graph')
plt.show()
```

```python
plt.boxplot(df['mpg'])
plt.ylabel('mpg')
plt.title('box plot')
plt.show()
```
